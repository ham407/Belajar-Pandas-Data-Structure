## Series

Series merupakan data struktur yang berbentuk array 1 dimensi yang berisi nilai. Pada dasarnya, series ini mirip dengan list dan array. Tapi, series ini memiliki banyak fungsi yang sangat berguna dalam melakukan data manipulation dan data analitycs.

Gambar di bawah ini merupakan struktur pada series.Pada bagian kiri itu merupakan index dari series, sedangkan bagian kanan merupakan value(nilai) dari series.




index.jpg

### Membuat series

|kode_barang|jumlah|
|:--|--:|
|IDMF|5|
|TROF|2|
|GODF|1|
|DETN|1|

Data di atas merupakan data transaksi barang. Misalnya, kita akan membuat data series dari kolom jumlah pembelian dari suatu transaksi. Untuk membuat series, dapat dilakukan dengan cara memasukkan nilai list pada fungsi `pd.Series()`.

```python
import pandas as pd
```

```python
# buat list data menggunakan python list
list_data = [5,2,1,1]
```

```python
# ubah data list menjadi data series
series_data = pd.Series(list_data)
```

```python
# tampilkan data series
series_data
```

Index akan secara otomatis mulai dari 0 hingga n (banyaknya data) jika index tidak diinput secara manual. Jika ingin index sesuai keiginan kita, dapat dilakukan melalui parameter index. Perlu diperhatikan bahwa panjang data harus sama. Misalnya jika panjang listnya ada 4, panjang indexnya juga harus 4.

```python
# buat list data dan index menggunakan python list
list_data = [5,2,1,1]
index_data = ["IDMF", "TROF", "GODF", "DETN"]
```

```python
# ubah data list menjadi data series
series_data = pd.Series(list_data, index=index_data)
```

```python
# tampilkan data series
series_data
```

selain dapat mengubah data list/array ke dalam bentuk series, data dictionary juga dapat diubah menjadi data series. Bahkan, lebih mudah dibandingkan list/array. Sebelumnya, kita harus memiliki 2 list (list untuk value dan list untuk index).

Tapi, dengan data dictionary, kita bisa langsung mengubahnya menjadi series. Index dan value dari dictionary akan otomatis menjadi index dan value data series.

### **Update data series**

Nilai data dapat diubah dengan cara mengakses index. Dapat dilakukan dengan menggunakan pola `namaseries[index] = nilaiupdate`

### **Tambah Data Baru**

Selanjutnya, ketika ingin menambahkan data baru ke dalam series, bisa menggunakan fungsi `append()`. Data terlebih dahulu dibuat dalam series baru, kemudian dimasukkan ke dalam fungsi. Setelah itu, series harus ditimpa karena hasil return dari fungsi ini tidak langsung menimpa series yang sudah ada sebelumnya. Contohnya `namaseries = namaseries.append(new_data)`

```python
# add data penjualan PGIN dan SGIN
# buat series baru
new_data = pd.Series([5,1], index = ["PGIN","SGIN"])

# tambahkan series baru ke series yang sudah ada sebelumnya
series_data = series_data.append(new_data)

# tampilkan data
series_data
```

```python
# delete 1 data, data transaksi SGIN
series_data = series_data.drop("SGIN")

# tampilkan data
series_data
```

```python
# delete 2 data atau lebih data, data transaksi PGIN dan IDMF
series_data = series_data.drop(["PGIN", "IDMF"])

# tampilkan data
series_data
```

### **Akses DataFrame**

Pada dataframe, terdapat **baris**, **kolom**, dan **cell**. Masing-masing punya cara mengakses yang berbeda-beda. Secara general kita bisa mengakses data dengan cara

* `dataframe.loc[[namaindex], [namakolom]]`

#### ***Akses Baris***

Mengakses baris pada dataframe dapat dilakukan  dengan cara:

* `dataframe.loc[[namaindex]]`

Pada fungsi `loc`, parameternya menggunakan nama index. Misalnya data Good Day itu nama indexnya "GODF".

```python
# akses baris menggunakan fungsi loc
df2.loc[["GODF"]]
```

```python
# akses lebih dari 1 baris menggunakan fungsi loc
df2.loc[["GODF","DETN"]]
```

#### ***Akses Kolom***

Mengakses kolom pada dataframe dapat dilakukan  dengan cara:

* `dataframe.loc[:,[namakolom]]`

Fungsi `loc` membutuhkan *nama kolom*.

Hal lain yang menjadi pembeda yaitu tanda `:`. Tanda itu berarti "select semua baris".

```python
# akses kolom menggunakan fungsi loc
df2.loc[:,["nama_barang"]]
```

```python
# akses lebih dari 1 kolom menggunakan fungsi loc
df2.loc[:,["nama_barang","jumlah"]]
```

#### ***Akses Cell***

Mengakses cell pada dataframe dapat dilakukan  dengan cara:

* `dataframe.loc[[namaindex],[namakolom]]`

Fungsi `loc` membutuhkan nama kolom/baris.

```python
# akses cell menggunakan fungsi loc
df2.loc[["GODF"],["nama_barang"]]
```

```python
# akses lebih dari 1 cell menggunakan fungsi loc
df2.loc[["GODF","DETN"],["nama_barang","jumlah"]]
```

### **Tambah data pada DataFrame**

Pada dataframe, kita dapat menambah baris atau kolom

* Menambah baris: `dataframe = dataframe.append(newdata)`
* Menambah kolom: `dataframe[namakolom] = newdata`

#### ***Menambah Baris***

```python
# menambah baris
# buat baris dalam bentuk dataframe

new_row = pd.DataFrame({
    "nama_barang": "Pasta Gigi",
    "jenis_barang": "non-food",
    "jumlah": 4},
    index=["PGIN"]
)

# tambahkan data baru ke dataframe yang telah dibuat sebelumnya
df2 = df2.append(new_row)

# tampilkan data
df2
```

### **Hapus data pada DataFrame**

Pada dataframe, kita dapat menghapus baris atau kolom menggunakan fungsi `drop()`

* Menghapus baris: `dataframe = dataframe.drop(namaindex)`
* Menghapus kolom: `dataframe = dataframe.drop(namaindex, axis=1)`

#### *Menghapus Baris*

```python
# menghapus baris
# menghapus baris pada index PGIN
df2 = df2.drop(["PGIN"])

# tampilkan data
df2
```

#### *Menghapus Kolom*

```python
# menghapus kolom harga_barang
df2 = df2.drop(["harga_barang"], axis=1)

# tampilkan data
df2
```

Parameter ini digunakan pada saat di dalam file csv tidak terdapat judul kolom.

## **Casting Tipe Data**

Pandas akan otomatis membaca jenis data dalam setiap kolom yang dibuat. Tapi, dalam melakukan analisis, kita kadang harus mengubah jenis data dalam kolom tersebut. Misalnya, ingin mengubah data bilangan integer menjadi bilangan decimal. Proses ini dinamakan "casting". Casting ini dapat dengan mudah dilakukan menggunakan pandas. Pandas memiliki fungsi `astype()` yang bisa digunakan untuk melakukan casting.

Misalnya, pada data di bawah ini, saya ingin mengubah data pada kolom "jumlah" dari integer (int64) menjadi decimal (float64)

```python
import pandas as pd
```

```python
# buat data dalam bentuk dictionary
data = {
    "nama_barang": ["Indomie","Taro","Good Day", "Detergen"],
    "jenis_barang": ["food","food","food","non-food"],
    "jumlah": [5, 2, 1, 1]
}

# buat dataframe dari data dictionary
df = pd.DataFrame(data)

# tampilkan tipe data tiap kolom
df.dtypes
```

```python
# casting kolom jumlah dari data integer ke float
df["jumlah"] = df["jumlah"].astype(float)

# tampilkan tipe data tiap kolom
df.dtypes
```

Contoh lain, saya ingin mengubah data "jenis_barang" dari object menjadi category.

```python
# casting kolom jenis_barang dari data object ke category
df["jenis_barang"] = df["jenis_barang"].astype("category")

# tampilkan tipe data tiap kolom
df.dtypes
```

## **Combining Data**

Sering kali data yang sudah tersedia dalam bentuk file yang terpisah. Kita sudah belajar tentang cara membaca data dari file menjadi dataframe. Tapi, ketika ingin mengolah data yang terpisah secara bersamaan, datanya harus digabung terlebih dahulu. maka dari itu, kita akan belajar cara mengombinasikan data.

Ketika kita memiliki 2 data yang berbeda sumbernya tetapi ingin diolah bersama, kita bisa menggabungnya menggunakan fungsi pada pandas. Misalnya data1 dari file *data1.csv* dan data2 dari *data2.csv*. Untuk menggabungkan data dapat dilakukan dengan cara:

### ***Concat***

Jika memiliki lebih dari 1 data dan ingin digabung, kita bisa menggunakan fungsi `concat()`. Fungsi ini dapat digunakan untuk data series dan dataframe.

Penggabungan data series sama dengan dataframe. Data dapat digabung dalam kolom yang sama.

<img src="https://drive.google.com/uc?export=view&id=1LnAQkz-WQT5FACWVTdrfrDpy9wiVwTx8" alt="Drawing" width= 500px;/>

```python
import pandas as pd
```

```python
# 2 dataframe dengan nama kolom yang BEDA

# buat dataframe1
transaksi_januari = pd.DataFrame({
    "nama_barang": ["Indomie", "Taro"],
    "jumlah": [5,2]
})

# buat dataframe2
transaksi_februari = pd.DataFrame({
    "nama_barang": ["Good Day", "Deterjen", "Pasta gigi"],
    "jumlah": [1,2,1]
})
```

```python
# gabungkan dataframe menggunakan fungsi concat
df_concat = pd.concat([transaksi_januari, transaksi_februari])
```

```python
# tampilkan data
df_concat
```

```python
# 2 dataframe dengan nama kolom yang SAMA
# buat dataframe1
transaksi_januari = pd.DataFrame({
    "nama_barang": ["Indomie", "Taro"],
    "jumlah": [5, 2]
    })

# buat dataframe2
stock_januari = pd.DataFrame({
    "nama_barang": ["Indomie", "Taro", "Pasta gigi"],
    "stock": [17,11,5]
    })

# gabungkan dataframe menggunakan fungsi concat
df_concat = pd.concat([transaksi_januari,stock_januari], axis=1)

# tampilkan data
df_concat
```

### ***Merge***


Cara lain untuk menggabung data yaitu dengan menggunakan fungsi `merge()`. Penggabungan dengan fungsi `.merge()` mirip seperti Klausa Join pada SQL.

Merge merupakan operasi penggabungan dataset dengan menghubungkan antarbaris menggunakan satu atau lebih *keys*. Data akan otomatis digabung dan disesuaikan pada kolom yang memiliki nilai yang sama.

<img src="https://drive.google.com/uc?export=view&id=1_r24EvYE1HkmObzAP1iTq1-1VVyRZnYO" alt="Drawing" width= 500px;/>

```python
# buat dataframe1
transaksi_januari = pd.DataFrame({
    "nama_barang": ["Indomie", "Taro"],
    "jumlah": [5, 2]
    })

# buat dataframe2
stock_januari = pd.DataFrame({
    "nama_barang": ["Indomie", "Taro", "Pasta gigi"],
    "stock": [17,11,5]
    })

# gabungkan data menggunakan fungsi merge
pd.merge(transaksi_januari, stock_januari, on="nama_barang")
```

Tapi, pada hasil penggabungan data di atas, data key pasta gigi hilang. Hasil penggabungan dataframe diatas merupakan inner join. Secara default, fungsi merge akan melakukan inner join terhadap data yang akan digabung. Selain inner join, ada left, right dan outer join. Untuk menerapkan jenis join data tersebut, kita bisa menggunakan parameter `how` pada fungsi `merge()`. Nilai dari parameter `how` dapat berupa `inner`, `left`, `right` atau `outer`.

```python
# gabungkan data menggunakan parameter how="outer"
pd.merge(transaksi_januari, stock_januari, on="nama_barang", how="outer")
```

Selain menggunakan fungsi `merge()`, kita juga dapat menggunakan fungsi `join()` untuk menggabung data. Namun, perlu diperhatikan bahwa data key pada dataframe harus menjadi index dari dataframe agar bisa tergabung. Secara default, fungsi `join()` akan menggabungkan dataframe dengan metode left join. Jika ingin melakukan metode join yang lain, bisa menggunakan parameter `how` seperti pada fungsi `merge()`.

```python
# buat dataframe1
transaksi_januari = pd.DataFrame({
    "nama_barang": ["Indomie", "Taro"],
    "jumlah": [5, 2]
    })
transaksi_januari.set_index('nama_barang', inplace=True)
# buat dataframe2
stock_januari = pd.DataFrame({
    "nama_barang": ["Indomie", "Taro", "Pasta gigi"],
    "stock": [17,11,5]
    })
stock_januari.set_index('nama_barang', inplace=True)
# gabungkan data menggunakan fungsi join
transaksi_januari.join(stock_januari)
```

```python
# gabungkan data menggunakan fungsi join dan parameter how="right"
transaksi_januari.join(stock_januari, how="right")
```
